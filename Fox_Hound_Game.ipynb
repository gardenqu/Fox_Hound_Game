```python
import matplotlib.pyplot as plt
import matplotlib.transforms as transforms

# create a 6x6 board
fig = plt.figure(figsize=[6,6])
fig.patch.set_facecolor((1,1,.8))
ax = fig.add_subplot(111)

# build a 45° rotation transform (around board center)
t = transforms.Affine2D().rotate_deg_around(3, 3, 45) + ax.transData
# Center is (3,3) since board runs 0..6

# draw the grid with the transform applied
for x in range(7):  # 0..6
    ax.plot([x, x], [0,6], 'k', transform=t)
for y in range(7):  # 0..6
    ax.plot([0,6], [y, y], 'k', transform=t)

# scale the axis area to fill the whole figure
ax.set_position([0,0,1,1])
ax.set_axis_off()
ax.set_xlim(-4, 10)   # widen limits so rotated board fits
ax.set_ylim(-4, 10)



plt.show()
```
